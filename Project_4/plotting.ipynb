# Using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(np.linspace(0,10000,10000))

y = np.array([(lambda num: 0.1*np.log(1+(10000/num)))(num) for num in x])

w = np.array([(lambda num: 0.2*(np.log(((num+10)/10000)+7)))(num) for num in x])

x1 = np.array(np.linspace(0,1, 10000))
z = np.array([(lambda num: np.log(num + 0.007) * 0.15 + 1)(num) for num in x1])
```

```python
plt.plot(x,y)
plt.savefig('./wordle_env/plots/fig1.png')
plt.show()
```

```python
plt.plot(w)
plt.savefig('./wordle_env/plots/fig2.png')
plt.show()
```

```python
plt.plot(z)
plt.savefig('./wordle_env/plots/fig3.png')
plt.show()
```
